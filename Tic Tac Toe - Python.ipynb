```python


import random
import time

"""
Tic Tac Toe Game
This is a non-OOP based Python program to implement a Tic-Tac-Toe game in Python.
Includes additional features like scoring and time limit.
"""



# Function to check whether the board is filled or not.
def is_board_full(board_ground):
    for row in range(3):
        for col in range(3):
            if board_ground[row][col] == '-':
                return False
    return True


# Function to check whether a player has won or not.
def check_for_winner(board_ground, player):
    for row in range(3):
        if board_ground[row][0] == board_ground[row][1] == board_ground[row][2] == player:
            return True

    for col in range(3):
        if board_ground[0][col] == board_ground[1][col] == board_ground[2][col] == player:
            return True

    if board_ground[0][0] == board_ground[1][1] == board_ground[2][2] == player:
        return True

    if board_ground[0][2] == board_ground[1][1] == board_ground[2][0] == player:
        return True

    return False


# Function to show the board.
def show_board(board_ground):
    print("\n")
    for row in range(3):
        for col in range(3):
            print(board_ground[row][col], end=" ")
        print("\n")



# Additional feature: Scoring and time limit
# initialize the score of each player to 0
scores = {'X': 0, 'O': 0}

# initialize the time limit to 10 seconds for each player
TIME_LIMIT = 10


# Function to start the game.
def play_game():
    # Select the first turn of the player randomly.
    print("Welcome to Tic Tac Toe Game!")
    player1 = input("Enter name for Player 1 (X): ")
    player2 = input("Enter name for Player 2 (O): ")
    # time delay of 3 seconds with countdown
    # Create a 3x3 board using a 2-dimensional array and initialize each element as empty.
    board = [['-', '-', '-'], ['-', '-', '-'], ['-', '-', '-']]
    scores['X'] = 0
    scores['Y'] = 0

    for i in range(3, 0, -1):
        print(i)
        time.sleep(1)
    players = {'X': player1, 'O': player2}
    current_player = random.choice(['X', 'O'])
    print("Current player is", players[current_player])

    while True:
        # Show the board to the user to select the spot for the next move.
        show_board(board)

        # Ask the user to enter the row and column number within the time limit.
        print(f"{players[current_player]}, you have {TIME_LIMIT} seconds to make your move.")
        start_time = time.time()
        while time.time() - start_time < TIME_LIMIT:
            row = input(f"Enter row number (0, 1, 2): ")
            col = input(f"Enter column number (0, 1, 2): ")
            if not row.isdigit() or not col.isdigit():
                print("Invalid input. Please enter valid integers.")
                continue
            row, col = int(row), int(col)
            if not (0 <= row <= 2) or not (0 <= col <= 2):
                print("Invalid input. Please enter row and column numbers between 0 and 2.")
                continue
            if board[row][col] != '-':
                print("Illegal move! This spot is already occupied.")
                continue
            break
        else:
            print("Time's up! You didn't make a move in time.")
            row, col = None, None

        # If the player did not make a move in time, skip their turn.
        if row is None or col is None:
            # Switch to the next player.
            if current_player == 'X':
                current_player = 'O'
            else:
                current_player = 'X'
            continue

        # Update the spot with the respective player sign.
        board[row][col] = current_player

        # Check whether the current player won the game or not.
        if check_for_winner(board, current_player):
            print(f"Player {current_player} has won the game!")
            scores[current_player] += 3
            break

            # Next, check whether the board is filled or not.
        if is_board_full(board):
            print("Game is drawn!")
            scores['X'] += 1
            scores['O'] += 1
            break

            # Switch to the next player.
        if current_player == 'X':
            current_player = 'O'
        else:
            current_player = 'X'

    # Finally, show the user the final view of the board.
    show_board(board)
    print("----Scoreboard----")

    print(f"Current scores: {players['X']}: {scores['X']}, {players['O']}: {scores['O']}")



# Call the play_game() function to start the game.
if __name__ == '__main__':
    play = True

    while play:
        play_game()
        print("----End----")
        play_again = input("Do you want to play again? (Y/N): ")
        if play_again.lower() != 'y':
            play = False






```
